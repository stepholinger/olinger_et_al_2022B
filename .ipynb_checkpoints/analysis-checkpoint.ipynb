```python
'''

Measure e-folding time and Q value from observations

'''

# set station and channel
station = "PIG2"
channel = "HHZ"

# read and process seismic data
st = obspy.read("data/MSEED/no_IR/" + station + "/" + channel + "/2012-05-09." + station + "." + channel + ".no_IR" + ".MSEED")
st.filter("highpass",freq=0.0005)
st.trim(starttime=obspy.UTCDateTime(2012, 5, 9, 17, 55))

# calculate peak amplitude
peak_amp = np.max(np.abs(st[0].data))
peak_idx = np.argmax(np.abs(st[0].data))

# calculate 1/e of peak amplitude
e_fold_amp = peak_amp * 1/np.e

# find peaks in signal that exceed e-folding amplitude
peaks = find_peaks(np.abs(st[0].data),height=e_fold_amp)

# get e-folding time
e_fold_time = st[0].times()[peaks[0][-1]] - st[0].times()[peak_idx]

# estimate period of signal during e-folding time
data = st[0].data[peak_idx:peaks[0][-1]] 
f = rfftfreq(len(data), 1/st[0].stats.sampling_rate)
spectra = abs(rfft(data))
T = 1/f[np.argmax(spectra)]

# calculate Q from e-folding time
num_cycles = e_fold_time/T
Q = num_cycles*np.pi
print("Estimated Q value: " + str(np.round(Q,2)))

# make plot showing e-folding values
t = st[0].times()
fig,ax = plt.subplots(figsize=(20,15))
ax.plot(t,st[0].data,'k')
ax.hlines(y=[-e_fold_amp,e_fold_amp],xmin=t[0],xmax=t[-1],color='r',linestyle='--')
ax.vlines(x=t[peaks[0][-1]],ymin=-peak_amp*1.1,ymax=peak_amp*1.1,color='r',linestyle='--')
ax.scatter(t[peaks[0]],st[0].data[peaks[0]],color='r')
ax.scatter(t[peak_idx],-peak_amp,s=1000,facecolor='gold',edgecolor='k',marker="*",zorder=3)
ax.text(200,-0.00002-peak_amp,"$A_{max}$ = " + str(np.round(peak_amp,5)) + " m/s",size=15)
ax.text(t[peaks[0][-1]],st[0].data[peaks[0][-1]]-0.000015,"   $\dfrac{1}{e}A_{max}$",size=15)
ax.text(t[peaks[0][-1]],-0.00002-peak_amp,r" $\tau = $" + str(np.round(e_fold_time/60)) + " minutes",size=15)
ax.set_ylim(-peak_amp*1.1,peak_amp*1.1)
ax.set_xlim(t[0],t[-1])
ticks = np.array([300, 3900, 7500, 11100, 14700, 18300, 21900])
ax.set_xticks(ticks)
ax.set_xticklabels([st[0].stats.starttime.datetime + datetime.timedelta(seconds=int(tick)) for tick in ticks])
ax.set_title("E-folding time for May 9 event",size=25)
ax.set_ylabel("Velocity (m/s)",size=15)
ax.set_xlabel("Time",size=15)
plt.show()
```
